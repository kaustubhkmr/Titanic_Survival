```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
x1 = [5,8,10]
y1 = [10,1,20]
style.use('ggplot')
plt.plot(x1,y1, label = 'value 1', linewidth = 5)

x2 = [5,2,12]
y2 = [6,1,10]
plt.plot(x2,y2, label = 'value 2')


plt.title('demo lines')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.axis([0, 15, 0, 25])
plt.show()

    # if you dont want to plot, but want to scatter
plt.scatter(x1,y1)
plt.show()

    # if you want to show the data in bars
plt.bar(x1,y1, color = 'g')
plt.show()

t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0., 5., 0.1)
t2 = np.arange(0., 5., 0.02)
plt.subplot(421)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.subplot(424)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.show()
```

# Images in matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img=mpimg.imread('stinkbug.png')
plt.imshow(img)
plt.show()
```

```python
lum_img = img[:,:,0]
plt.imshow(lum_img)
plt.show()
```

```python
plt.imshow(lum_img, cmap="hot")
plt.show()
```

```python
plt.imshow(lum_img, cmap="nipy_spectral")
plt.show()
```

```python
plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.show()
```

```python
plt.imshow(lum_img, clim=(0.0, 0.7))
plt.show()
```

```python
from PIL import Image
img = Image.open('stinkbug.png')
img.thumbnail((64, 64), Image.ANTIALIAS) # resizes image in-place
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img, interpolation="nearest")
plt.show()
```

```python
plt.imshow(img, interpolation="bicubic")
plt.show()
```
